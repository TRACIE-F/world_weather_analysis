```python
from citipy import citipy

# Dependencies
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from config import weather_api_key
import time
```

```python
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
coordinates = list(lat_lngs)
coordinates[0:10]
```

```python
cities = []

for coord in coordinates:
    lat = coord[0]
    long = coord[1]
    
    city = citipy.nearest_city(lat, long)
    city_name = city.city_name
    
    if city_name not in cities:
        cities.append(city_name)
        
print(len(cities))
```

```python
# init lists
city_longs = []
city_lats = []
city_descriptions = []
city_temps_max = []
city_humids = []
city_clouds = []
city_winds = []
cities_df = []
countries = []

weather_url = "http://api.openweathermap.org/data/2.5/weather"
```

```python
for city in cities[0:50]:
    print(city)
    query_url = f"{weather_url}?appid={weather_api_key}&q={city}&units=imperial"
    
    response = requests.get(query_url)

    if (response.status_code == 200):
        weather_data = response.json()
        
        try:
            long = weather_data["coord"]["lon"]
            lat = weather_data["coord"]["lat"]
            description = weather_data["weather"][0]["description"]
            temp_max = weather_data["main"]["temp_max"]
            cloud = weather_data["clouds"]["all"]
            humid = weather_data["main"]["humidity"]
            wind = weather_data["wind"]["speed"]
            city_p = weather_data["name"]
            country = weather_data["sys"]["country"]

            # append to lists
            city_longs.append(long)
            city_lats.append(lat)
            city_descriptions.append(description)
            city_temps_max.append(temp_max)
            city_clouds.append(cloud)
            city_humids.append(humid)
            city_winds.append(wind)
            cities_df.append(city_p)
            countries.append(country)
        
        except Exception as e:
            print(e)
            
            city_longs.append(np.nan)
            city_lats.append(np.nan)
            city_descriptions.append(np.nan)
            city_temps_max.append(np.nan)
            city_clouds.append(np.nan)
            city_humids.append(np.nan)
            city_winds.append(np.nan)
            cities_df.append(city)
            countries.append("")
    else:
        print(response.status_code)
        print(response.text)
        
    time.sleep(1)
```

```python
# Dataframe
random_df = pd.DataFrame()
random_df["City"] = cities_df
random_df["Country"] = countries
random_df["Max Temp"] = city_temps_max
random_df["Humidity"] = city_humids
random_df["Cloudiness"] = city_clouds
random_df["Wind Speed"] = city_winds
random_df["Latitude"] = city_lats
random_df["Longitude"] = city_longs
random_df["Current Description"] = city_descriptions

random_df
```

```python
random_df.to_csv("WeatherPy_Database.csv", index=False) 
```
